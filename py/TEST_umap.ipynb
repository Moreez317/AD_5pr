```python
import sklearn.datasets
from sklearn import preprocessing
import pandas as pd
import numpy as np
import umap.umap_ as umap
import umap.plot as umapPlot
import matplotlib.pyplot as plt
import ssl
```

```python
ssl._create_default_https_context = ssl._create_unverified_context

pendigits = sklearn.datasets.load_digits()
mnist = sklearn.datasets.fetch_openml('mnist_784')
```

```python
#mapper = umap.UMAP().fit(pendigits.data)

n_n = (5, 25, 50)
m_d = (0.1, 0.6)

um = dict()
for i in range(len(n_n)):
    for j in range(len(m_d)):
        um[(n_n[i], m_d[j])]=(umap.UMAP(n_neighbors=n_n[i], min_dist=m_d[j], random_state=123).fit(pendigits.data))
```

```python
umapPlot.points(um[5, 0.1], labels=pendigits.target)
```
